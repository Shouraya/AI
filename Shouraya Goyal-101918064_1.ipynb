```python
# Question 1

'''
Finding the sum of the series:
(1 + x + (x2)/2! + . . . + (xn)/n!)

Number x and n should be entered at runtime.
'''

# x = the base value of the series
# n = the number of terms upto which the sum of the series has to be found

# Function to find the factorial
def factorial(num):
    fact=1
    
    for i in range(1, num+1):
        fact=fact*i
        
    return fact    


# Function to find the Sum of a series
def Series(x, n):
    sum = 1

    for i in range(1, n):
        sum = sum + ((x) ** i)/(factorial(i))
        
    return sum


# Taking inputs from the end user
x = int(input('Enter the base value of your series: '))
n = int(input('Enter the number of terms upto which the sum of the series has to be found: '))

print(Series(x,n))
```

```python
# Question 2

'''
WAP to create a list of 100 random numbers between 100 and 900. Count and print the:

(i) All odd numbers
(ii) All even numbers
(iii) All prime numbers
'''

import random

r = random.sample(range(100, 900), 100)

print(r)
print('\nrandomList\'s Size -> ', len(r))

even = []
odd = []
prime = [] 

for currentIndex in range(0, (len(r))):
    
    currentElement = r[currentIndex]

    if (currentElement % 2) == 0:
        even.append(currentElement)
    else:
        if currentElement > 1:  
            if (currentElement == 2 or currentElement == 3):
                prime.append(currentElement)
            
            interimNumbersList = range(2 , ((currentElement//2)+1))
            
            '''
            Refreshing for every number [Every number has to traverse the array of factors in interimNumberslist ..
            .. before they are declared as a prime number by their non-divisibility by any one of those factors.
            '''
            
            count = 0
            
            for i in interimNumbersList:
                if (currentElement % i) == 0:
                    odd.append(currentElement)
                    break
                else:
                    count += 1
                    
                    '''
                    Since I can't call it a prime number directly before the divisibility is checked by all the 
                    interim factors.
                    '''
                    
                    if(count >= len(interimNumbersList)):
                        prime.append(currentElement)
                        
print ('Total Odd Numbers -> ', len(odd))
print ('Total Even Numbers -> ', len(even))
print ('Total Prime Numbers -> ', len(prime))
```

```python
# Question 3

'''
Find all the prime numbers between two given numbers.
'''

def generatePrimeList(number1, number2):
    
    outputList = []
    
    # Assumption Made - Extremes excluded.
    for num in range(number1+1, number2):
        
        if num > 1:  
            if (num == 2 or num == 3):
                outputList.append(num)
            
            interimNumbersList = range(2 , ((num//2)+1))
            
            '''
            Refreshing for every number [Every number has to traverse the array of factors in interimNumberslist ..
            .. before they are declared as a prime number by their non-divisibility by any one of those factors.
            '''
            
            count = 0
            
            for i in interimNumbersList:
                if (num % i) == 0:
                    break
                else:
                    count += 1
                    
                    '''
                    Since I can't call it a prime number directly before the divisibility is checked by all the 
                    interim factors.
                    '''
                    
                    if(count >= len(interimNumbersList)):
                        outputList.append(num)
                        
                        
    return outputList

number1 = int(input('Enter the first number of your Range: '))
number2 = int(input('Enter the second number of your Range: '))

output = generatePrimeList(number1, number2)

print('\nThe list of Prime Numbers between ', number1, ' & ', number2, ' is:')
print(output)

print('\nThe Total Number of Prime Numbers between ', number1, ' & ', number2, ' are -> ', len(output))
```

```python
# Question 4

'''
Find the common elements from two lists
'''

# Function to return the common elements in the list passed to it.
def commonElementsList( list1 , list2 ):
    
    setA = set(list1)
    setB = set(list2)
 
    if (setA & setB):
        output = list(setA & setB)
    else:
        output = "No common elements"
        
    return output
          

print(commonElementsList(["ramesh",3,4], [4,"Ramesh",6]))
```

```python
# Question 5

'''
Write a Python Program to input basic salary of an employee and calculate its Gross
salary according to following: Basic Salary <= 10000 : HRA = 20%, DA = 80% Basic
Salary <= 20000 : HRA = 25%, DA = 90% Basic Salary > 20000 : HRA = 30%, DA =
95%.
'''

# Making a function to generalize the calculative process
def grossSalaryCalculator(salary):
    
    hra = 0
    da = 0 
    
    if (salary <= 10000):
        hra = 0.2
        da = 0.8
    elif (salary <= 20000):
        hra = 0.25
        da = 0.9
    elif (salary > 20000):
        hra = 0.3
        da = 0.95
        
    return (salary + (salary * hra) + (salary * da))
        
    
    
employeeSalary = int(input("Enter the basic salary of Employee -> "))
print('The gross salary after hra\'s & da\'s allowances is ->' , grossSalaryCalculator(employeeSalary))
```

```python
# Question 9

'''
D is a dictionary defined as D= {1:”One”, 2:”Two”, 3:”Three”, 4: “Four”, 5:”Five”}.
(i) WAP to add new entry in D; key=6 and value is “Six”
(ii) WAP to remove key=2.
(iii) WAP to check if 6 key is present in D.
(iv) WAP to count the number of elements present in D.
(v) WAP to add all the values present in D.
'''

D = {
    1 : "One",
    2 : "Two",
    3 : "Three",
    4 : "Four",
    5 : "Five"
}

# (i)
D[6] = "Six"
print(D)

# (ii)
del D[2]
print(D)

# (iii)
if (6 in D.keys()):
    print('\nVoila! It\'s Present.')
    
# (iv)
print('\nThere are total' , len(D.keys()) , 'elements in our dictionary.')

# (v)

from word2number import w2n
literalValuesList =[]

valuesList = D.values()
for values in valuesList:
    transformedValue = w2n.word_to_num(values)
    literalValuesList.append(transformedValue)

# Now adding all these transformed values
print('Total of all the values in our python dictionary is -> ', sum(literalValuesList))

# Even if it's about adding the strings, here's the solution
finalString = ""
for stringValue in valuesList:
    finalString += stringValue + ' '
    
print(finalString)
```
